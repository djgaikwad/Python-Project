#### Import Modules

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from selenium import webdriver
from time import sleep
from webdriver_manager.chrome import ChromeDriverManager
```

#### Set Path For Webdriver

```python
path= 'C//chromedriver.exe'
```

```python
# Open the browser
browser=webdriver.Chrome(executable_path=path)
```

```python
# Load the webpage
browser.get('https://www.flipkart.com')
browser.maximize_window()
```

```python
browser.find_element("xpath",'//button[@class="_2KpZ6l _2doB4z"]').click()
Event = 'iphone'
browser.find_element("xpath",'//input[@class="_3704LK"]').send_keys(Event)
browser.find_element("xpath",'//button[@class="L0Z3Pu"]').click() 
```

```python
product_name=[]
product_rate=[]
cust_rating=[]
no_of_ratings=[]
no_of_reviews=[]
image_link= []

for i in range(17):
    print('scraping page',i+1)
    
    
    image = browser.find_elements("xpath",'//img[@class="_396cs4"]')
    for j in image[:24]:
        image_link.append(j.get_attribute('src'))
        
    product = browser.find_elements("xpath",'//div[@class="_4rR01T"]')
    for p in product[:24]:
        product_name.append(p.text)
        
    rate = browser.find_elements("xpath",'//div[@class="_30jeq3 _1_WHN1"]')
    for r in rate[:24]:
        product_rate.append(r.text)   
    
    rating= browser.find_elements("xpath",'//div[@class="_3LWZlK"]')
    for q in rating[:24]:
        cust_rating.append(q.text)
    
    no= browser.find_elements("xpath",'//span[@class="_2_R_DZ"]')
    for n in no[:24]:
        n1=n.text.split(' ')
        no_of_ratings.append(n1[0])
        no_of_reviews.append(n1[3]) 
    browser.get(('https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page={x}').format(x=i+2))
     
    sleep(4)    
```

```python
#coverting scrape into CSV
import pandas as pd
dict = {"Product":product_name,"Price":product_rate,'Rating':cust_rating,'No. of Ratings':no_of_ratings,
        'No. of Reviews':no_of_reviews,'images':image_link}
df = pd.DataFrame(dict)
```

```python
df
```

```python
df.info()
```

```python
df['Price']=df['Price'].str.replace('₹','')
df['Price']=df['Price'].str.replace(',','')
df['No. of Ratings']=df['No. of Ratings'].str.replace(',','')
df['No. of Reviews']=df['No. of Reviews'].str.replace(',','')
```

```python
df['Price']=df['Price'].astype(int)
df['Rating']=df['Rating'].astype(float)
df['No. of Ratings']=df['No. of Ratings'].astype(int)
df['No. of Reviews']=df['No. of Reviews'].astype(int)
```
